```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}
plt.rc('font', **font)
```

```python
scores_per_lunarlander = np.load("../models/LunarLander-v2_episodes_2000_score_58.686178929696622019_05_18_23_32_40per_scores.npy")
scores_lunarlander = np.load("../models/LunarLander-v2_episodes_2000_score_145.619876247867672019_05_18_23_31_21regularbuffer_scores.npy")
N = 20
scores_per_lunarlander_ave = np.convolve(scores_per_lunarlander, np.ones((N,))/N, mode='valid')
scores_lunarlander_ave = np.convolve(scores_lunarlander, np.ones((N,))/N, mode='valid')
plt.figure(figsize=(10, 8))
plt.plot(scores_per_lunarlander_ave, label="per")
plt.plot(scores_lunarlander_ave, label="reg")
plt.legend()
```
