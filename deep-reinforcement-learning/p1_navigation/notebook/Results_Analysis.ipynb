```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}
plt.rc('font', **font)
```

## Regular Replay buffer vs Prioritized Replay Buffer

```python
## LunarLander 
scores_per_lunarlander = np.load("../models/LunarLander-v2_episodes_1229_score_201.35206195665882019_05_19_16_02_30__scores.npy")
scores_lunarlander = np.load("../models/LunarLander-v2_episodes_2000_score_118.34985029313562019_05_19_00_59_58_reg_buffer_size_2000_scores.npy")
N = 100
scores_per_lunarlander_ave = np.convolve(scores_per_lunarlander, np.ones((N,))/N, mode='valid')
scores_lunarlander_ave = np.convolve(scores_lunarlander, np.ones((N,))/N, mode='valid')
plt.figure(figsize=(10, 8))
plt.plot(scores_per_lunarlander_ave, label="PriortizedReplay")
plt.plot(scores_lunarlander_ave, label="Replay")
plt.xlabel("Num of episodes")
plt.ylabel("Scores (Moving Average of 100)")
plt.legend()
plt.savefig("../models//LunarLander-v2_scores.png")
```

```python
## LunarLander 
scores_per_banana = np.load("../models/Banana_unity_episodes_1182_score_13.012019_05_19_17_50_40_banana_per_scores.npy")
scores_banana = np.load("../models/Banana_unity_episodes_1263_score_13.02019_05_19_17_21_56_banana_reg_scores.npy")
N = 100
scores_per_ave = np.convolve(scores_per_banana, np.ones((N,))/N, mode='valid')
scores_ave = np.convolve(scores_banana, np.ones((N,))/N, mode='valid')
plt.figure(figsize=(10, 8))
plt.plot(scores_per_ave, label="PriortizedReplay")
plt.plot(scores_ave, label="Replay")
plt.xlabel("Num of episodes")
plt.ylabel("Scores (Moving Average of 100)")
plt.legend()
plt.savefig("../models/Banana_unity_scores.png")
```
